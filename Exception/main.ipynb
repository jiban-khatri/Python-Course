##### Exceptions
If things go wrong during the execution of the program(runtime). It generally happens when something unforeseen has happened.<br>
* Exceptions are raised by python runtime.
* You have to tackle is on the fly
##### Examples
* Memory overflow
* Divide by 0 -> logical error
* Database error

```python
# how to handle exception
# ---> Try except block
```

```python
class Bank:
        
    def __init__(self,balance):
        self.balance = balance

    def withdraw(self,amt):
        if amt<0:
            raise Exception ("Money can't  be -ve")
        if self.balance<amt:
            raise Exception ("You don't have enough money")
        
        self.balance -= amt
    

b = Bank(12000)

try:
    b.withdraw(5000)
except Exception as e:
    print(e)
else:
    print("Remaining Balance : ",b.balance)
finally:
    print("Money Withdrawn successfully")
```
