Writing to a file

```python
f = open("sample.txt", "w")
f.write("Hello")
f.close()
```

```python
# write many content
L = ["Hi\n", "Hello\n", "Jiban\n", "Khatri\n", "How are you"]
f = open("sample.txt", "w")
f.writelines(L)
f.close()
```

```python
# Reading from Files

# -> using read() ---- read all at once
f = open("sample.txt", 'r')
s = f.read()
f.close()

print(s)
```

```python
#  reading upto n characters
f = open("sample.txt", 'r')
s = f.read(10)  # write no. of characters to read
f.close()

print(s)
```

```python
# readline() -- read line by line
f = open("sample.txt", 'r')
print(f.readline())
print(f.readline())
f.close()
```

```python
# CONTEX MANAGER (With)

with open("sample.txt", "w") as f:
    f.write("Writing with W")

```

```python
with open("sample.txt","r") as f:
    print(f.readline())

```
