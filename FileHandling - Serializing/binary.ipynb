Serialization and Deserialization<br>
* Serialization - process of converting python data types to JSON format
* Deserialization - process of converting JSON to python data types

What is JSON?<br>
* JavaScript Object Notation
* same as python dictionaries
<br>


````python
# ----> Serialization using JSON module <----
# List

import json

L = [1,2,3,4,5]

with open("demo.json","w") as f:
    json.dump(L,f)
````

````python
# dictionary
d = {
    "name" : "jiban",
    "age" : 18
}
with open("demo.json","w") as f:
    json.dump(d,f,indent=4) # indent for better representation
````

````python
# deserialization

with open("demo.json","r") as f:
    d2 = json.load(f)
    print(d2)
    print(type(d2))

````

````python
# serialize and deserialize tuple
T = (1,2,3,4,5)

with open("demo.json","w") as f:
    json.dump(T,f)      # tuple converts to list
````

````python
# serializing and deserializing custom objects

class Person:

    def __init__(self,fname, lname, age, gender):
        self.fname = fname
        self.lname = lname
        self.age = age
        self.gender = gender

p = Person("Jiban", "Khatri", 18,"Male")




import json

def show_object(p):
    if isinstance(p,Person):
        return f"First Name : {p.fname}, Last Name : {p.lname}, Age : {p.age}, Gender : {p.gender}"

with open("demo2.json","w") as f:
    json.dump(p,f,default=show_object)
````

````python
with open("demo2.json","r") as f:
    print(json.load(f))
````

##### Pickling
```Pickling``` is the process whereby a Python object hierarchy is converted into a byte stream and<br>
```unpickling``` is the inverse operation, whereby a byte stream (from a binary file or bytes like object) is <br>
converted back into an object hierarchy.

````python
class Person:

    def __init__(self,name,age):
        self.name = name
        self.age = age
    
    def display_info(self):
        print(f"Hi my name is {self.name} and my age is {self.age}.")


p = Person("Jiban",18)
````

````python
# pickle dump

import pickle

with open("person.pkl","wb") as f:
    pickle.dump(p,f)
````

````python
# pickle load 

with open("person.pkl","rb") as f:
    p = pickle.load(f)

p.display_info()
````

Pickle Vs JSON<br>
* Pickle lets the user to store data in binary format. JSON lets the user store data in a human readable text format
