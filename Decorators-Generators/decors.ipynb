##### Decorators 

* Takes function as input and returns function as output but with extended functionality ( basically decorates the functionand returns )
* Types
    * Built in  : ``` @staticmethod, @classmethod ```
    * User Defined


````python
def decorator(func):
    def wrapper():
        print("*********")
        func()
        print("*********")
    return wrapper

def greet():
    print("Hello Jiban")

a = decorator(greet)
a()
````

````python
# Better syntax

def my_decorator(func):
    def wrapper():
        print("*********")
        func()
        print("*********")
    return wrapper

@my_decorator
def greet():
    print("Hello Jiban")

greet()
````

````python
# decorator that shows the execution time to execute a function

import time

def timer(func):
    def wrapper():
        start = time.time()
        func()
        print("Time taken by",func.__name__, (time.time() - start),"sec")  # func.__name__ returns function name
        
    return wrapper

@timer
def hello():
    print("hello world")
    time.sleep(2)       # delays time for 2 sec

hello()
````

````python
# decorator that checks that the input is valid or not

def sanity_check(data_type):
    def outer_wrapper(func):
        def inner_wrapper(*args):
            if(type(*args)) == data_type:
                func(*args)
            else:
                raise TypeError("Yo datatype chaldaina")
        return  inner_wrapper
    return outer_wrapper

@sanity_check(int)
def square(num):
    print(num**2)

square(4)
square("h") # error
````
