Write OOP Classes to handle the following scenarios:<br>
- A user can create and view 2d coordinates<br>
- A user can find out the distance between two coordinates<br>
- A user can find the distance of a coordinate from origin<br>
- A user can check if a point lies in a given line<br>
- A user can find the distance between a given 2D points and a given line<br>

```python

import math

class Point:

    def __init__(self,x,y):
        self.x_cod = x
        self.y_cod = y


    def __str__(self):
        return "({},{})".format(self.x_cod, self.y_cod)

    def calc_dist(self,other):
        distance = math.sqrt( (other.x_cod - self.x_cod)**2 + (other.y_cod - self.y_cod)**2)
        return ('The distance is {}'.format(distance))

    def dist_from_origin(self):
        distance = math.sqrt( (self.x_cod)**2 + (self.y_cod)**2)
        return ('The distance is {}'.format(distance))


# Line Eqn and Coord
class Line:
    def __init__(self,A,B,C):
        self.A = A
        self.B = B
        self.C = C
    
    def __str__(self):
        return "{}x + {}y + {} = 0".format(self.A, self.B, self.C)

    def point_on_line(line,point):
        if line.A*point.x_cod + line.B*point.y_cod + line.C == 0:
            return "The points ({},{}) lie in the equation {}x + {}y + {} = 0".format(point.x_cod, point.y_cod, line.A, line.B, line.C)
        else:
            return "The points ({},{}) doesn't lie in the equation {}x + {}y + {} = 0".format(point.x_cod, point.y_cod, line.A, line.B, line.C)

    def dist_point_line(line, point):
        distance = abs(line.A*point.x_cod + line.B*point.y_cod + line.C) / math.sqrt(line.A**2 + line.B**2)
        return "The distance between ({},{}) and the equation {}x + {}y + {} = 0 is {}".format(point.x_cod, point.y_cod, line.A, line.B, line.C, distance) 


l1 = Line(2,3,5)
p1 = Point(1,1)

print("Eqn of Line is {}".format(l1))
print("Given Points = {}".format(p1))

print(l1.point_on_line(p1))
print(l1.dist_point_line(p1))
```

```python
# Refrence Variable = Holds the object

class Person:
    def __init__(self,name):
        self.name = name
    
    def greet(self):
        return ("Namaste, {}".format(self.name))

p = Person("Jiban") # Here p is the refrence variable
print(p.greet())       

q = p
q.name = "Ram"

print(p.name)
print(id(p))

print(q.name)
print(id(q))
```
