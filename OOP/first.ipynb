```python
# ATM Machine Code using OOP

class Atm:
    # constructor - basically a function inside a class. It is a special function. not needed to call the constructior function to call constructor. When object is created then the code automatically executes

    def __init__(self):
        self.pin = ''
        self.balance = 0
        self.menu()


    def menu(self):
        user_input = input("""
        "Hi! How can i help you?
        1. Press 1 to Create pin.
        2. Press 2 to change pin.
        3. Press 3 to check balance.
        4. Press 4 to withdraw.
        5. Anything else to exit.
        """)

        if user_input == "1":
            self.create_pin()
        elif user_input == '2':
            self.change_pin()
        elif user_input == '3':
            self.check_balance()
        elif user_input == '4':
            self.withdraw_balance()
        else:
            self.atm_exit()
    
    def create_pin(self):
        user_pin = input("Create Pin: ")
        self.pin = user_pin

        user_balance = int(input("Enter your balance"))
        self.balance = user_balance

        print("Pin created successfully")
        self.menu()

    def change_pin(self):
        old = input("Enter your current Pin")
        if self.pin == old:
            new_pin = input("Enter new pin: ")
            self.pin = new_pin
            print("Pin changed successfully")      
        else:
            print("Sorry cannot change")

        self.menu()
    
    def check_balance(self):
        user_pin = input("Enter your pin:")
        if user_pin == self.pin:
            print(f"Your balance is {self.balance}")
        else:
            print("Get Lost")
        
        self.menu()
    
    def withdraw_balance(self):
        user_bal = int(input("Enter balance to withdraw: "))
        if self.balance >= user_bal:
            user_pin = input("Enter pin to withdraw:")
            if user_pin == self.pin:
                print(f"{user_bal} withdraw success")
                self.balance -= user_bal
            else:
                print("Sorry pin doesn't match")
        else:
            print("Sorry amount exceeds")
        
        self.menu() 
    
    def atm_exit(self):
        print("Thank you for using ATM")
        exit()


atm = Atm()     
        
```

```python
# Own Fracction DataType

class Fraction:

    def __init__(self,x,y):
        self.num = x
        self.den = y
        
    def __str__(self):
        return f"{self.num}/{self.den}"

    def __add__(self, other):
        new_num = self.num*other.den + other.num*self.den
        new_den = self.den * other.den
        return f"{new_num}/{new_den}"
    
    def __sub__(self, other):
        new_num = self.num*other.den - other.num*self.den
        new_den = self.den * other.den
        return f"{new_num}/{new_den}"
    
    def __mul__(self, other):
        new_num = self.num * other.num
        new_den = self.den * other.den
        return f"{new_num}/{new_den}"

    def __truediv__(self, other):
        new_num = self.num*other.den 
        new_den = self.den * other.num
        return f"{new_num}/{new_den}"



fr1 = Fraction(1,2)
fr2 = Fraction(1,4)

print(fr1 + fr2)
print(fr1 - fr2)
print(fr1 * fr2)
print(fr1 / fr2)

```

```python
# Temperature Converter

class Temperature:
    def __init__(self):
        self.initial_temp = float("0")
        self.changed_temp = float("0")
        self.menu()


    def menu(self):
        init_temp_method = input("Enter Temperature Mode(C, F, K)").lower()
        user_temp = float(input("Enter temperature:"))
        change_method = input("Enter method to be changed (C, F, K)").lower()
        
        if init_temp_method == "c" and change_method == 'f':
            self.initial_temp = user_temp
            self.cel_fer()

        elif init_temp_method == "c" and change_method == 'k':
            self.initial_temp = user_temp
            self.cel_kel()

        elif init_temp_method == "f" and change_method == 'c':
            self.initial_temp = user_temp
            self.fer_cel()

        elif init_temp_method == "f" and change_method == 'k':
            self.initial_temp = user_temp
            self.fer_kel()

        elif init_temp_method == 'k' and change_method == 'c':
            self.initial_temp = user_temp
            self.kel_cel()

        elif init_temp_method == 'k' and change_method == 'f':
            self.initial_temp = user_temp
            self.kel_fer()
        
    def cel_fer(self):
        self.changed_temp = (self.initial_temp * (9/5)) +32
        print(f"{self.initial_temp} Celsius in Fahrenheit is {round(self.changed_temp)}")

    def cel_kel(self):
        self.changed_temp = self.initial_temp + 273.15
        print(f"{self.initial_temp} Celsius in Kelvin is {round(self.changed_temp)}")

    def fer_cel(self):
        self.changed_temp = (self.initial_temp - 32) * (5/9)
        print(f"{self.initial_temp} Fahrenheit in Celsius is {round(self.changed_temp)}")

    def fer_kel(self):
        self.changed_temp = (self.initial_temp - 32) * (5/9) + 273.15
        print(f"{self.initial_temp} Fahrenheit in Kelvin is {round(self.changed_temp)}")

    def kel_cel(self):
        self.changed_temp = self.initial_temp - 273.15
        print(f"{self.initial_temp} Kelvin in Celsius is {round(self.changed_temp)}")

    def kel_fer(self):
        self.changed_temp = (self.initial_temp - 273.15) * (9/5) + 32
        print(f"{self.initial_temp} Kelvin in Fahrenheit is {round(self.changed_temp)}")


temp = Temperature()
```

```python
class Rectangle:
    def __init__(self,l,w):
        self.len = l
        self.wd = w
        self.pm = 2 *(self.len + self.wd)
        self.area = self.len * self.wd


    def display(self):
        print("The Length is ",self.len)
        print("The Width is ",self.wd)
        print("The Perimether is ",self.pm)
        print("The Area is ",self.area)


my_rectangle = Rectangle(3,4)
my_rectangle.display()
```
