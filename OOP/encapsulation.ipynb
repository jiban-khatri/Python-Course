```python
# Encapsulation - Hiding private data inside an object. Use __name (double underscore) to make something private. Expose it only through methods.

class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner           # public  
        self.__balance = balance     # private 
        self.__pin = 1234            # private


    def deposit(self, amount):
        print(f"Deposit {amount}")
        self.__balance += amount

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrawn {amount}")
        else:
            print("Insufficient funds!")

    # GETTER — a safe way to READ the balance
    def get_balance(self):
        return self.__balance



acc = BankAccount("Alice", 5000)

acc.deposit(500)       
acc.withdraw(200)          
  

# Try to hack the balance directly:
acc.__balance = 99999  #  - Not Possible
print(acc.get_balance())    # Prints 5300
```

```python
# List of Objects

class Person:
    def __init__(self,name,gender):
        self.name = name
        self.gender = gender
    
    def __str__(self):
        return f"{self.name, self.gender}"

p1 = Person("Jiban","Male")
print(p1)
```
