```python
#  Aggregation ( Has-A relationship)

class Customer:
    def __init__(self,name,gender,address):
        self.name = name
        self.gender = gender
        self.address = address
    
    def print_cust_details(self):
        print(f"Name = {self.name}")
        print(f"Gender = {self.gender}")
        print(f"Address = {self.address.get_city(), self.address.pin, self.address.country}")

    def change_profile(self, new_name, new_city,new_pin, new_country):
        self.name = new_name
        self.address.change_new_address(new_city,new_pin, new_country)



class Address:
    def __init__(self,city,pin,country):
        self.__city = city
        self.pin = pin
        self.country = country

    def change_new_address(self,new_city, new_pin, new_country):
        self.__city = new_city
        self.pin = new_pin
        self.country = new_country


    def get_city(self):
        return self.__city



add = Address("Kathmandu", 44600, "Nepal")
cust = Customer("Jiban", "Male", add)
cust.print_cust_details()


cust.change_profile("Umesh", "Tulsipur", 22400, "Nepal")
cust.print_cust_details()

```

```python
# Inheritance

# parent class
class User:
    def __init__(self):
        self.name = "jiban"
    
    def login(self):
        print("Login")


# child class
class Student(User):
    
    def enroll(self):
        print("Enroll into class")

u = User()
s = Student()
print(s.name)
s.enroll()
s.login()
```

```python
# constructor example

class Phone:
    def __init__(self,pname,pos):
        print("Inside Phone Constructor")
        self.name = pname
        self.os = pos

    def buy(self):
        print("Buying Phone")

class SmartPhone(Phone):
    pass


s = SmartPhone("Iphone","IOS")
print(s.name)
s.buy()
```

```python
# child can't access private members of parent

class Phone:
    def __init__(self,pname,pos):
        print("Inside Phone Constructor")
        self.name = pname
        self.__os = pos

    def buy(self):
        print("Buying Phone")

class SmartPhone(Phone):
    def show_os(self):
        print(self.__os)        # cant access private members


s = SmartPhone("Iphone","IOS")
print(s.name)
s.show_os()
```

```python
# super

class Phone:
    def __init__(self,pname,pos):
        print("Inside Phone Constructor")
        self.name = pname
        self.__os = pos

    def buy(self):
        print("Buying Phone")

class SmartPhone(Phone):
    def buy(self):
        print("Buying a smartPhone")
        # syntax to call parent method
        super().buy()        


s = SmartPhone("Iphone","IOS")
s.buy()
```

```python
class Phone:
    def __init__(self,price,brand,camera):
        print("Inside Phone Constructor")
        self.__price = price
        self.brand = brand
        self.cam = camera

    def buy(self):
        print("Buying Phone")

class SmartPhone(Phone):
    def __init__(self, price,brand,camera, os,ram):
        super().__init__(price,brand,camera)
        self.os = os
        self.ram = ram
        print("Inside Smartphone constructor")
    

s = SmartPhone(2000,"Apple",10,"IOS",8)
print(s.os)
print(s.brand)
```
