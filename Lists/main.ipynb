# Python Lists

- list is a datatype where we can store multiple items under 1 variable name<br>
Eg : ``` L = [1,2,3,4,5] ```<br><br>

### Characteristics <br>
- Ordered <br>
- Mutable/Changable <br>
- Hetrogeneous<br>
- Can have duplicates <br>
- Are Dynamic <br>
- can be nested <br>
- Items can be accessed <br>
- Can contain of any kind of objects

# List VS Array

- Array are pre Fixed in size || Lists are Dynamic, flexible <br>
- Arrays are homogeneous ( stores only same type of data) || Lists are hetrogeneous ( can store data of any datatype) <br>
- Arrays are Faster || Lists are slower than arrays <br>
- Arrays takes less memory || Lists take more memory

Creating a List

````python
# Empty List
print( [] )

#  1D list -> Homogeneous
print([1,2,3,4,5])

# 2D List -> Nested Lists
print([1,2,3,[4,5]])

# 3D List
print([ [[1,2],[3,4]] , [[4,5],[6,7]] ])

#  Hetrogeneous
print([1,2,'a',4.6,3+2j])

# using type conversion
print(list('hello'))
````

Accessing Items from Lists

````python
# ***  Indexing -> starts from 0 ***
l = [1,4,7,2,8]
# positive indexing
print(l[0])
# negative indexing
print(l[-1])

l2 = [1,2,3,[4,5]]
print(l2[3][0])
print(l2[-1][-1])



# *** Slicing ***
L = [1,2,4,5,6,3,7]
print(L[0:4])

print(L[-2:-6:-1])
````

Adding items to a list

````python
# append -> adds item at the last position of list -> can only add at a time

l = [1,2,3,5,3]
l.append(7)
print(l)
````

````python
#  extend -> can add multiple at a time

l = [1,2,3,4,5]
l.extend([6,7,8])
print(l)
````

````python
#  insert -> adds item at desired position

l = [1,32,54,6,3,37,12,1]

l.insert(4,100)
print(l)
````

Edit existing item in list

````python
l = [1,2,4,6,5]
l[3] = 29
print(l)

l[1:3] = [14,43]
print(l)
````

Deleting items from list

````python
#  del

l = [1,2,4,1,4]
print(l)

# del l -- deletes whole list
del l[3]
print(l)
````

````python
# remove

l = [1,2,4,1,5]

l.remove(4) # removes item valued 4
print(l)
````

````python
# Pop 

l = [1,2,4,1,4]

l.pop() #removes items from last
print(l)

l.pop(0)    # removes item of index 0
print(l)
````

````python
# clear

l = [1,2,4,1,4]
l.clear()  #clears the list deleting all items but it doesn't delete whole list

print(l)

````

#### Operations on List <br>
- Arithmetic <br>
- Membership <br>
- Loop

````python
# Arithmetic Operations

l1 = [1,2,3]
l2 = [3,4,5]

print(l1 + l2)  #merge two or more lists

print(l1*2)  #repeat 2 times and merge
````

````python
#  Membership Operators

l1 = [1,2,3,4,5,6]
l2 = [4,5,6,7,8]

print(3 in l1)
print(l2 in l1)

print(8 not in l2)
````

````python
#  Loop

l1 = [1,2,3,4,5,6]
l2 = [4,5,6,[7,8]]

# for i in l1:
#     print(i)

# for i in l2:
#     print(i)

l3 = [ [[1,2],[3,4]] , [[4,5],[6,7]] ]

for i in l3:
    print(i)
````
