## List Comprehension

##### List comprehension provides a consise way of creating list <br>
```newlist = [ expression for item in iterable if condition == True]```

#### Advantages: <br>
- More time efficient and space efficient than loop <br>
- require fewer lines of code <br>
- transforms iterative statement into a formula

````python
# Add 1 to 10 items in a list

'''l = []
for i in range(1,11):   #Long Method
    l.append(i)

print(l)'''

#  Shorter:
 
l = [i for i in range(1,11)]
print(l)
````

````python
#  Scalar multiplication on vector

'''v = [2,3,4]
s = -2
n = []
for i in v:
    n.append(i*s)
print(n)'''


v = [2,3,4]
s = -2

n = [s*i for i in v]
print(n)
````

````python
#  Add squares 

L = [1,2,3,4,5]
L = [i**2 for i in L]
print(L)
````

````python
# Print all numbers divisible by 5 in range of 1-50

A = [i for i in range(1,51) if i%5==0]
print(A)
````

````python
# find languages which starts with letter 'p'

languages = ['java','python','php','c','javascript']

print([each for each in languages if each.startswith('p')])
````
