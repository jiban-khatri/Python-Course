#### Dictionary

Dictionary is a collection of keys, values used to store different types of values<br>
- ``` dict = { 'name':'jiban', 'age':18 }```

Characteristics:<br>
- Mutable<br>
- Indexing has no meaning<br>
- keys can't be duplicate<br>
- keys can't be mutable items<br>

````python
# empty dictionary

my_dict = {}
print(my_dict)

# 1D dictionary
d = {'name' : 'jiban', 'gender' : 'male'}
print(d)

# with mixed keys
d1 = {(1,2,3) : 1, 'hello' : 'world'}
print(d1)

#2d dictionary
d2 = {
    'name' : 'jiban', 
    'address' : { 
        'p_address' : 'Dang',
        't_address' : 'KTM' 
        }
    }
print(d2)


# Dictionary with dict function

d4 = dict([('name','jiban'), ('age',18)])
print(d4)
````

````python
# duplicate keys are not allowed

d = {
    'name' : "jiban",
    "name" : "hari"
}
print(d)
````

````python
# mutable items are not allowed
d = {
    [1,3] : (2,3)   #not allowed as list are mutable
}
print(d)
````

````python
# == Accessing items ==

'''my_dict = {
    "name" : "jiban",
    'gender' : 'male',
    'age' : 18
}

print(my_dict['age'])
print(my_dict.get('gender'))'''

d2 = {
    'name' : 'jiban', 
    'address' : { 
        'p_address' : 'Dang',
        't_address' : 'KTM' 
        }
    }
print(d2['address']['p_address'])

d2['continent'] = {}
d2['continent']['country'] = 'Nepal'
print(d2)
````

````python
#  -- Adding key value pair --

my_dict = {
    "name" : "jiban",
    'gender' : 'male',
    'age' : 18
}

my_dict['country'] = 'Nepal'
print(my_dict)
````

````python
#  == Editing Key value ==

my_dict = {
    "name" : "jiban",
    'gender' : 'male',
    'age' : 18,
    'country' : 'Nepal'
}

my_dict['age'] = 19
print(my_dict)
````

````python
# ----- Operations -------

# 1. Membership
d2 = {
    'name' : 'jiban',
    'gender' : 'Male',
    'age' : 18, 
    'address' : { 
        'p_address' : 'Dang',
        't_address' : 'KTM' 
        }
    }

# 'jiban' in my_dict      # false becaues in dict everything is about key
'name' in my_dict
````

````python
d = {
    'name' : 'jiban',
    'gender' : 'Male',
    'age' : 18, 
    'address' : { 
        'p_address' : 'Dang',
        't_address' : 'KTM' 
        }
    }

for each in d:
    print(f"{each} = {d[each]}")
````

````python
# == Methods ==

d = {
    'name' : 'jiban',
    'gender' : 'Male',
    'age' : 18, 
    'address' : { 
        'p_address' : 'Dang',
        't_address' : 'KTM' 
        }
    }

print(len(d))
print(sorted(d))    # sorts the keys in ascending order
````

````python
#  items
print(d.items())    #returns keys and values in tuple form inside list

# keys
print(d.keys())

# Values
print(d.values())
````

##### Dictionary Comprehension

````python
#  Print first 10 numbers with their squares

{ i : i**2 for i in range(1,11) }
````

````python
#  dictionary using existing dictionary
d = {'ktm' : 1000, 'bkt' : 2000, 'ltp' : 3000}
{ i : d[i]*.62 for i in d}
````

````python
# using zip
days = ['sunday','monday','tuesday','wednesday','friday']
temp = [30.5, 32, 37, 28, 40, 33]

dict(zip(days,temp))
````
