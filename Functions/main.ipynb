<br>
```
def sum(a,b):       # function definition, name and parameters
```

````python
# program to check whether the number is even or odd

def is_even(n):
    if n % 2 == 0:
        return 'Even'
    else:
        return 'Odd'

print(is_even(3))
print(is_even(90))
````

*args and **kwargs<br>
*args and **kwargs are special python keywords that are used to pass the variable length of arguments to a function

````python
#  *args 
# allows us to pass a variable number of non keyword arguments to a function

def muliply(*args):
    product = 1
    for i in args:
        product = product * i

    return product   


muliply(1,2,3,4,5)
````

````python
#  **kwargs
#  **kwargs allows you to pass any number of keywoard arguments
#  *keyword arguments means that they contain a key value pair, like a python dictionary

def display(**kwargs):
    for (key,value) in kwargs.items():
        print(key,'->',value)

display(Nepal = "kathmandu", India = "Delhi")
````

##### Lambda Function<br>
A lambda function is a small anonymous function<br>
A lambda function can take any number of arguments, but can only have one expression<br>
``` 
lambda a,b : a+b
```
<br>
- No name<br>
- No return Value ( returns a function)<br>
- lambda is written in 1 line<br>
- Not reusable

````python
# x - x^2
square = lambda x : x**2
print(square(2))

#x,y -> x+y
sum  = lambda x,y : x+y
print(sum(2,3))
````

````python
#  check if a string has 'a'
check = lambda s : 'a' in s
print(check('jiban'))
````

````python
# check odd or even
checkNum = lambda num : "Even" if num%2 == 0 else "Odd"
checkNum(3)
````

````python
# Hgher order function : function that returns a function / function that accept other function

def square(x):
    return x**2

def transform(f,l):
    output = []
    for i in l:
        output.append(f(i))
    print(output)


l = [1,2,3,4,5]
transform(square,l)
````

````python
# 1) Map
l = [1,2,3,4,5]
list(map(lambda x : x**2, l))
````

````python
l = [1,2,3,4,5]
list(map(lambda x : 'even' if x%2==0 else 'odd', l))
````

````python
users = [
    {
        "name" : "Jiban",
        "age" : 18
    },
    {
        "name" : "Umesh",
        "age" : 18
    },
    {
        "name" : "Anup",
        "age" : 18
    },
    {
        "name" : "Sagar",
        "age" : 18
    }
]

list(map(lambda user : user["name"],users))
````

````python
#  Filter
l = [1,2,3,4,5,6,7]
list(filter(lambda x : x<5, l))
````

````python
fruits = ['apple', 'guava', 'cherry']
list(filter(lambda fruit : fruit.startswith('g'),fruits))
````

````python
#  Reduce

import functools

n = [2,3,1,4,5]
functools.reduce(lambda x,y : x+y, n)
````

````python
# find min
functools.reduce(lambda x,y : x if x<y else y, n)
````
