```python
# Program to find length of string wiithout using len method

s = input("Enter a string: ")
count = 0
for i in s:
    count += 1

print("Length of string is ",count)
```

```python
# count the frequency of a particular character in a provided string

msg = input("enter any msg: ")
term = input("What's frequency do you want to count: ")
print("Frequency: ",msg.count(term))
```

```python
#  WAP which can remove particular character from string

s = input("enter any string")
ch = input("Enter character to remove")

result = ''

for i in s:
    if i != ch:
        result += i

print(result)
```

```python
#  Program to check whether the given string is palindrome or not

s = input("Enter string")
reverse = ''

for ch in s:
    reverse = ch + reverse

if reverse == s:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
#  show words without split

s = input("Enter a sentence: ")
word = ""

for ch in s:
    if ch != " ":
        word += ch       
    else:
        print(word)       
        word = ""        

print(word)              
```

```python
#  convert given string into title case without title()

s = input("Enter any string: ")
 
splited = s.split()
titled = []

for i in splited:
    titled.append(i.capitalize())

print(' '.join(titled))
```
