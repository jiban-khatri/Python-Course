#### Tuple<br>
A tuple in pyhon is similar to lists

Characteristics:<br>
 - Ordered<br>
 - Immutable<br>
 - Allows Duplicate<br><br>
 Creating a tuple:<br>
 ``` T = (1,2,3,4,5,)```      

````python
t = (1,5,2,4,5,6)
print(t[0])

print(t[-1])
print(t[0:5])
print(t[-1:-6:-1])
````

````python
# Operations = '+' and '*' 

t1 = [2,3,1,1]
t2 = [3,9,1,4]
print(t1+t2)

print(t1 * 3)
````

````python
# Tuple Unpacking

a,b,c = (1,2,3)
print(a,b,c)
````

````python
a,b = (1,3,4)       #both side must be equal
print(a,b)
````

````python
# swapping values

a = 1
b = 2

a,b = b,a
print(a,b)
````

````python
a,b,*others = (1,2,3,4,5)

print(a)
print(b)
print(others)   
````
