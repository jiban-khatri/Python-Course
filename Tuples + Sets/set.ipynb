#### Set

Set are python datatype where items are unique. Items are not repeatable.<br><br>
Characteristics:<br>
- Unique Items<br>
- Unordered<br>
- Mutable but doesn't contain mutable datatype items<br>
<br>
##### Creaing Set<br>
``` 
    Empty Set:
        s = set()
    Set with items:
        s1 = {1,2,3,4,5} 
```

````python
s = set()
print(s)
print(type(s))
````

````python
s1 = {1,2,3,5}
print(s1)

s = {1,2,4,[3,5]}   #error cause list is mutable and set doesn't contain mutable item
s = {1,2,4,(3,5)}   #true no error
print(s)
````

````python
s = {1,2,3,1,3,5}
print(s)

print(s[0]) #Not Possible as set is unordered
````

````python
# Adding and removing 

s = {1,2,4}
s.add(3)    #can only add one item at once
print(s)

s.update([26,12])   #can add multiple items at once
print(s)

# del s         deletes the set
# print(s)    

s.discard(78)     #removes item from set but if item is not exist in set then it doesn't throw error
print(s)

s.remove(2)    #removes item from set but if item is not exist in set then throws error
print(s)

s.pop()     # removes randomly any item
print(s)

s.clear()   # clears the set
print(s)
````

````python
#  Operations
s1 = {1,2,3,4,5}
s2 = {4,5,6,7}

# 1. Union( | )
s1 | s2

# 2. Intersection( & )
s1 & s2

# 3. Difference( - ) => (only)
s1-s2

# 4. Symmetric difference( ^ ) => returns all values expect intersection
s1 ^ s2


#  Membership
1 in s1
1 not in s1

#  Iteration
for i in s1:
    print(i)
````

````python
#  Methods

# len,min,max,sum,sorted
s = {1,2,3,4,5}
len(s)
min(s)
max(s)
sum(s)
sorted(s)
sorted(s,reverse=True)
````

````python
s1 = {1,2,3,4,5}
s2 = {4,5,6,7}

# Union 
s1.union(s2)

# intersection
s1.intersection(s2) 
````
